# BRCAssure

## Extraction of text from pdf and save in seperate text file [BRCAssure]

```python
import os
import pdfminer
from pdfminer.high_level import extract_text

folder_path = "D:\lab report classification\BRCAssure"
output_folder = "D:\lab report classification\BRCAssure_text"
os.makedirs(output_folder, exist_ok=True)
for filename in os.listdir(folder_path):
    if filename.endswith('.pdf'):
        pdf_path = os.path.join(folder_path, filename)
        text = extract_text(pdf_path)
        output_file_pdf = os.path.join(output_folder, os.path.splitext(filename)[0] + ".txt")
        with open(output_file_pdf, 'w', encoding='utf-8') as file:
            file.write(text)
        print("Text extracted from", filename, "and saved to", output_file_pdf)
```

## Data preprocessing and store it in a list [BRCAssure]

```python
import os
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def preprocess_text(text):
 
    text = re.sub('[^a-zA-Z0-9]', ' ', text)
    tokens = word_tokenize(text)
    
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    clean_text = ' '.join(tokens)
    return clean_text.lower()


folder_path = "D:\lab report classification\BRCAssure_text"

text_brc = []

for file_name in os.listdir(folder_path):
    if file_name.endswith('.txt'):
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()

        preprocessed_text = preprocess_text(text)
        text_brc.append(preprocessed_text)

print(text_brc)
```

## DataFrame [BRCAssure]

```python
import pandas as pd
df_brc = pd.DataFrame(text_brc,columns = ['text'])
```

```python
df_brc
```

```python
df_brc['label'] = 1  
```

# BLOOD TEST

# Naive Bayes

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics

x = final_df.drop('text', axis=1) 
y = final_df['label']
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

from sklearn.naive_bayes import GaussianNB


model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt="d", cmap="Spectral")

plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```
